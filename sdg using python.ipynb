```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn as sk
```

```python
df=pd.read_csv("/content/housepricess.csv")
df
```

```python
from sklearn import preprocessing
sx=preprocessing.MinMaxScaler()
sy=preprocessing.MinMaxScaler()
scaled_x=sx.fit_transform(df.drop('price',axis='columns'))
scaled_y=sy.fit_transform(df.price.values.reshape(df.shape[0],1))
```

```python
import random
```

```python
def sgd(x,y_true,epochs,learning_rate=0.5):
  nooffeatures=x.shape[1]
  w=np.ones(shape=nooffeatures)
  b=0
  total_samples=x.shape[0]

  epoch_list=[]
  cost_list=[]

  for i in range(epochs):
    random_index=random.randint(0,x.shape[0]-1)
    sample_x=x[random_index]
    sample_y=y_true[random_index]
    y_predicted=np.dot(w, sample_x.T)+b

    wd = -(2/total_samples)*(sample_x.T.dot(sample_y-y_predicted))
    bd = -(2/total_samples)*np.sum(sample_y-y_predicted)

    w=w-learning_rate*wd
    b=b-learning_rate*bd

    cost=np.mean(np.square(sample_y-y_predicted))

    if i%10==0:
      epoch_list.append(i)
      cost_list.append(cost)

  return w,b,cost,epoch_list,cost_list

wsgd,bsgd,cost, epoch_listsgd,cost_listsgd=sgd(scaled_x, scaled_y.reshape(scaled_y.shape[0]) ,200,learning_rate=0.5)
wsgd,bsgd,cost


```

```python
def predict(area,bedrooms,wsgd,bsgd):
  sampleofx=sx.transform([[area,bedrooms]])[0]
  scaled_price=np.dot(wsgd,sampleofx.T)+bsgd
  return sy.inverse_transform([[scaled_price]])[0][0]
predict(1440,3,wsgd,bsgd)
```
